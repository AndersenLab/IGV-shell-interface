```python
import pandas as pd
import cyvcf2
from cyvcf2 import VCF
import numpy as np
```

```python
def import_VCF(file = 'for_garth.vcf'):
    df = VCF(file)
    samples = df.samples

    result = pd.DataFrame()
    result.index.name = 'ID'

    for variant in df:
        sample_types = pd.Series(data=variant.gt_types, index=samples)
        variant_params = pd.Series({'CHROM':variant.CHROM, 'start':int(variant.start), 'end':int(variant.end), 'filter':variant.FILTER})
        row = pd.concat([variant_params, sample_types])
        row.name = variant.ID
        result = result.append(row)
        

    result['HOM_REF'] = result.apply(lambda x: x.index[np.where(x==0)[0][0]], axis=1) # Get the first strain (index order) that is HOM_REF
    result['HET'] = result.apply(lambda x: x.index[np.where(x==1)[0][0]], axis=1) # Get the first strain (index order) that is HET
    result['HOM_ALT'] = result.apply(lambda x: x.index[np.where(x==3)[0][0]], axis=1) # Get the first strain (index order) that is HOM_ALT
    result['start'] = result['start'].astype(int)
    result['end'] = result['end'].astype(int)
    result['length'] = result['end']-result['start']
    return(result)

df = VCF('for_garth.vcf')
```
